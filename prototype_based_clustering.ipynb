DBSCAN algorithm using DBC_Customers_1 dataset

```python
import pandas as pd
from sklearn.cluster import DBSCAN
import matplotlib.pyplot as plt

# Load the data
data = pd.read_csv('/content/drive/MyDrive/data mining program assignment/Copy of DBC_Customers_1.csv')

# Check the column names in your dataset
print(data.columns)

# Rename the columns to remove any extra spaces
data.columns = data.columns.str.strip()

# Select the features you want to use for clustering
# For example, let's use 'Annual Income ($)' and 'Spending Score (1-100)'
X = data[['Annual Income ($)', 'Spending Score (1-100)']]

# Define the DBSCAN model with appropriate parameters
epsilon = 5000  # Adjust this value based on your data and requirements
min_samples = 5  # Adjust this value based on your data and requirements
dbscan = DBSCAN(eps=epsilon, min_samples=min_samples)

# Fit the model to the data
dbscan.fit(X)

# Get the labels assigned to each data point
labels = dbscan.labels_

# Visualize the clustering results (scatter plot)
plt.scatter(X['Annual Income ($)'], X['Spending Score (1-100)'], c=labels, cmap='rainbow')
plt.xlabel('Annual Income ($)')
plt.ylabel('Spending Score (1-100)')
plt.title('DBSCAN Clustering Results')
plt.show()
```

```python
# Import necessary libraries
import pandas as pd
from sklearn.cluster import SpectralClustering
from sklearn.preprocessing import StandardScaler
import matplotlib.pyplot as plt

# Load the dataset
data = pd.read_csv('/content/drive/MyDrive/data mining program assignment/Copy of DBC_Customers_1.csv')  # Replace 'your_dataset.csv' with the actual file path

# Select the features for clustering (e.g., Annual Income and Spending Score)
features = data[['Annual Income ($)', 'Spending Score (1-100)']]

# Standardize the data (important for Spectral Clustering)
scaler = StandardScaler()
scaled_features = scaler.fit_transform(features)

# Create and fit the Spectral Clustering model
spectral_clustering = SpectralClustering(n_clusters=3, affinity='rbf', random_state=0)  # You can adjust the parameters as needed
cluster_labels = spectral_clustering.fit_predict(scaled_features)

# Add cluster labels to the original dataset
data['Cluster'] = cluster_labels

# Visualize the clusters
plt.scatter(data['Annual Income ($)'], data['Spending Score (1-100)'], c=cluster_labels)
plt.xlabel('Annual Income ($)')
plt.ylabel('Spending Score (1-100)')
plt.title('Spectral Clustering')
plt.show()

# Now, 'Cluster' contains the cluster assignments for each data point
# You can explore and analyze the clusters in your dataset

# Example: Count the number of data points in each cluster
cluster_counts = data['Cluster'].value_counts()
print(cluster_counts)
```

K MEANS algorithm on DBC_Customers_1 dataset

```python
import pandas as pd
from sklearn.cluster import KMeans
import matplotlib.pyplot as plt

# Load the dataset
data = pd.read_csv("/content/drive/MyDrive/data mining program assignment/Copy of DBC_Customers_1.csv")  # Replace "your_dataset.csv" with the actual file path

# Select the features for clustering
features = data[["Age", "Annual Income ($)", "Spending Score (1-100)"]]

# Choose the number of clusters (K) - you need to specify this
k = 3  # You can change this to the desired number of clusters

# Initialize and fit the K-means model
kmeans = KMeans(n_clusters=k, random_state=0)
data['Cluster'] = kmeans.fit_predict(features)

# Plot the clusters
colors = ['r', 'g', 'b', 'y', 'c', 'm']
for i in range(k):
    cluster_data = data[data['Cluster'] == i]
    plt.scatter(cluster_data['Annual Income ($)'], cluster_data['Spending Score (1-100)'], c=colors[i], label=f'Cluster {i}')

plt.scatter(kmeans.cluster_centers_[:, 1], kmeans.cluster_centers_[:, 2], s=300, c='k', marker='x', label='Centroids')
plt.title(f'K-means Clustering with K={k}')
plt.xlabel('Annual Income ($)')
plt.ylabel('Spending Score (1-100)')
plt.legend()
plt.show()
```

K-MEANS using HBC_timesData_2 dataset

```python
import pandas as pd
from sklearn.cluster import KMeans
import matplotlib.pyplot as plt
from sklearn.impute import SimpleImputer

# Load the university ranking dataset
data = pd.read_csv("/content/drive/MyDrive/data mining program assignment/Copy of HBC_timesData_2.csv")  # Replace with your actual file path

# Replace '-' with NaN
data = data.replace('-', float('nan'))

# Select the features for clustering
features = data[["teaching", "international", "research", "citations", "income"]]

# Impute missing values using mean values
imputer = SimpleImputer(strategy="mean")
features_imputed = imputer.fit_transform(features)

# Choose the number of clusters (K) - you need to specify this
k = 2 # You can change this to the desired number of clusters

# Initialize and fit the K-means model
kmeans = KMeans(n_clusters=k, random_state=0)
data['Cluster'] = kmeans.fit_predict(features_imputed)

# Plot the clusters
colors = ['r', 'g', 'b', 'y', 'c', 'm']
for i in range(k):
    cluster_data = data[data['Cluster'] == i]
    plt.scatter(cluster_data['teaching'], cluster_data['research'], c=colors[i], label=f'Cluster {i}')

plt.scatter(kmeans.cluster_centers_[:, 2], kmeans.cluster_centers_[:, 4], s=300, c='k', marker='x', label='Centroids')
plt.title(f'K-means Clustering with K={k}')
plt.xlabel('Teaching Score')
plt.ylabel('Research Score')
plt.legend()
plt.show()
```

```python
import pandas as pd
from sklearn.cluster import KMeans
import matplotlib.pyplot as plt

# Load the user data dataset
data = pd.read_csv("/content/drive/MyDrive/data mining program assignment/Copy of PBC_userprofile_3.csv")  # Replace with your actual file path

# Select the features for clustering (latitude and longitude)
features = data[["latitude", "longitude"]]

# Choose the number of clusters (K) - you need to specify this
k = 3  # You can change this to the desired number of clusters

# Initialize and fit the K-means model
kmeans = KMeans(n_clusters=k, random_state=0)
data['Cluster'] = kmeans.fit_predict(features)

# Plot the clusters on a map
colors = ['r', 'g', 'b', 'y', 'c', 'm']
for i in range(k):
    cluster_data = data[data['Cluster'] == i]
    plt.scatter(cluster_data['longitude'], cluster_data['latitude'], c=colors[i], label=f'Cluster {i}')

plt.scatter(kmeans.cluster_centers_[:, 1], kmeans.cluster_centers_[:, 0], s=300, c='k', marker='x', label='Centroids')
plt.title(f'K-means Clustering with K={k}')
plt.xlabel('Longitude')
plt.ylabel('Latitude')
plt.legend()
plt.show()
```
